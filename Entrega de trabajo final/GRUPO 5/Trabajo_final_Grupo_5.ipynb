Utilizando scipy. stats, realiza un análisis de una distribución normal con una media de
0 y desviación estándar de 1. Genera un conjunto de datos aleatorios con 1000
muestras a partir de esta distribución.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Parámetros de la distribución normal
media = 0
desviacion_estandar = 1
num_muestras = 1000

# Generar datos aleatorios a partir de una distribución normal
datos = np.random.normal(media, desviacion_estandar, num_muestras)
datos
```

Calcula y muestra la media, la desviación estándar y la mediana de los datos
generados.

```python
# Calcular estadísticas básicas
media_muestral = np.mean(datos)
desviacion_muestral = np.std(datos, ddof=1)  # ddof=1 para la desviación estándar muestral
mediana_muestral = np.median(datos)
print(f"La media es: {media_muestral}")
print(f"La desviación estándar es: {desviacion_muestral}")
print(f"La mediana es: {mediana_muestral}")
```

Grafica el histograma de los datos y superpón la curva de la distribución normal
teórica.

```python
# Ajuste de la distribución normal a los datos
parametros = stats.norm.fit(datos)

# Generar valores teóricos de la distribución normal
x = np.linspace(min(datos), max(datos), 1000)
pdf_teorica = stats.norm.pdf(x, *parametros)

# Graficar el histograma de los datos y la curva teórica de la distribución normal
plt.figure(figsize=(8, 5))
plt.hist(datos, bins=30, density=True, alpha=0.6, color='b', label="Histograma de datos")
plt.plot(x, pdf_teorica, 'r', label="Distribución normal teórica")
plt.xlabel("Valor")
plt.ylabel("Densidad de probabilidad")
plt.title("Análisis de distribución normal")
plt.legend()
plt.grid(True)
plt.show()
```

Calcula la probabilidad de que un valor caiga dentro del rango de -1 a 1 (una desviación
estándar de la media).

```python
# Calcular la probabilidad de que un valor caiga dentro de [-1, 1]
probabilidad = stats.norm.cdf(1, media, desviacion_estandar) - stats.norm.cdf(-1, media, desviacion_estandar)
probabilidad
```
