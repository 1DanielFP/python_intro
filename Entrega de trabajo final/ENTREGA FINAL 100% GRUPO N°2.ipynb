Utilizando scipy. stats, realiza un análisis de una distribución normal con una media de
0 y desviación estándar de 1. Genera un conjunto de datos aleatorios con 1000
muestras a partir de esta distribución.

```python
from scipy.stats import norm
import numpy as np

datos = norm.rvs(0, 1, 1000)

print(datos)  
```

Calcula y muestra la media, la desviación estándar y la mediana de los datos
generados.

```python
from scipy.stats import norm
import numpy as np

datos = norm.rvs(0, 1, 1000)

media= np.mean(datos)
print("Media:", media)

#DESVIACIÓN ESTÁNDAR:
desviacion_estandar=np.std(datos)
print("Desviacion estándar:",desviacion_estandar)

#MEDIANA:
mediana=np.median(datos)
print("Mediana:",mediana)
```

Grafica el histograma de los datos y superpón la curva de la distribución normal
teórica.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(40)  

hist = np.random.normal(0, 1, 1000)
plt.hist(hist, bins=20, density=True, edgecolor='black')
plt.plot(np.linspace(-4, 4, 100), norm.pdf(np.linspace(-4, 4, 100), 0, 1))
plt.show()
```

Calcula la probabilidad de que un valor caiga dentro del rango de -1 a 1 (una desviación
estándar de la media).

```python
from scipy.stats import norm

probab = norm.cdf(1) - norm.cdf(-1)
print("probabilidad de caer entre -1 y 1:", probab)
```

```python
# Indicadores de Desarrollo GlobaL, a nivel país:

import pandas as pd
import matplotlib.pyplot as plt

# Cargar el CSV usando delimitador punto y coma y codificación utf-8
file_path = r"C:\Users\ASUS\Documents\Python Scripts\desarrollo global.csv"
df = pd.read_csv(file_path, delimiter=";", encoding="utf-8")

# Filtrar los datos para Perú y el indicador "GDP growth (annual %)"
df_filtered = df[(df["Country Name"] == "Peru") & (df["Series Name"] == "GDP growth (annual %)")]

# Transformar los datos de formato ancho a largo (melt)
df_long = df_filtered.melt(id_vars=["Country Name", "Series Name"],
                           var_name="Year", value_name="Value")

# Extraer el año de la columna "Year" y manejar filas sin dato
df_long["Year"] = df_long["Year"].str.extract(r'(\d{4})')[0]
df_long = df_long.dropna(subset=["Year"])  # elimina filas sin un año válido
df_long["Year"] = df_long["Year"].astype(int)

# Convertir la columna "Value" a numérico y eliminar filas sin datos válidos
df_long["Value"] = pd.to_numeric(df_long["Value"], errors="coerce")
df_long = df_long.dropna(subset=["Value"])

# Graficar la evolución del crecimiento del PIB en Perú
plt.figure(figsize=(10, 5))
plt.plot(df_long["Year"], df_long["Value"], marker="o", linestyle="-", color="b")
plt.title("Evolución del crecimiento del PIB en Perú")
plt.xlabel("Año")
plt.ylabel("GDP growth (annual %)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el CSV usando punto y coma y la codificación adecuada
file_path = r"C:\Users\ASUS\Documents\Python Scripts\desarrollo global.csv"  # Ajusta la ruta si es necesario
df = pd.read_csv(file_path, delimiter=";", encoding="utf-8")

# Verificar las columnas disponibles (opcional)
print("Columnas:", df.columns.tolist())

# Filtrar datos para el indicador "CO2 emissions (metric tons per capita)"
df_co2 = df[df["Series Name"] == "CO2 emissions (metric tons per capita)"]

# Seleccionar solo el nombre del país y el valor para 2017
df_co2_2017 = df_co2[["Country Name", "2017 [YR2017]"]].copy()

# Convertir la columna de 2017 a numérico y eliminar filas sin datos
df_co2_2017["2017 [YR2017]"] = pd.to_numeric(df_co2_2017["2017 [YR2017]"], errors="coerce")
df_co2_2017 = df_co2_2017.dropna()

# Ordenar de mayor a menor y tomar los 10 primeros países
df_top10 = df_co2_2017.sort_values(by="2017 [YR2017]", ascending=False).head(10)

# Crear un gráfico de barras simple
plt.figure(figsize=(10, 6))
plt.bar(df_top10["Country Name"], df_top10["2017 [YR2017]"], color="purple")
plt.xlabel("País")
plt.ylabel("Emisiones de CO₂ (toneladas per cápita)")
plt.title("Top 10 países con mayores emisiones de CO₂ (2017)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
