Grafica la distribución binomial completa para los 10 lanzamientos, mostrando la
probabilidad de obtener desde 0 hasta 10 caras.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

x = np.arange(0, 11)

y = binom.pmf(x, 10, 0.5)

plt.bar(x,y, color= "green")

plt.xlabel("Número de caras")
plt.ylabel("Probabilidad")
plt.title("Distribución Binomial (n=10, p=0.5)")
plt.xticks(x)
plt.show()
```

Utilizando scipy. stats, realiza un análisis de una distribución normal con una media de
0 y desviación estándar de 1. Genera un conjunto de datos aleatorios con 1000
muestras a partir de esta distribución.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm 

datos = norm.rvs(0, 1, size=1000)

plt.hist(datos, bins=np.linspace(-4,4,30), edgecolor='black', density=True)
plt.xlim(-4,4)
plt.show()
```

Calcula y muestra la media, la desviación estándar y la mediana de los datos
generados.

```python
import numpy as np
import scipy.stats as stats

muestra = stats.norm.rvs(loc=0, scale=1, size=1000) 

resumen = stats.describe(muestra)

promedio = np.round(np.mean(muestra), 2)
desv_std = np.round(np.std(muestra, ddof=1), 2)
valor_medio = np.round(np.median(muestra), 2)

print("=== Resultados Estadísticos ===")
print(f"→ Promedio: {promedio}")
print(f"→ Desviación estándar: {desv_std}")
print(f"→ Valor central (Mediana): {valor_medio}")
```

Grafica el histograma de los datos y superpón la curva de la distribución normal
teórica.

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 1000) 
y = stats.norm.pdf(x, loc=0, scale=1)  

plt.figure(figsize=(8, 5))
plt.hist(datos, bins=30, density=True, alpha=0.6, color='b', label='Histograma de datos')
plt.plot(x, y, 'r', label='Densidad teórica')

plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.title("Histograma de datos simulados vs. Distribución Normal Teórica")
plt.legend()
plt.show()
```

```python
df = df.dropna(subset=['ubigeo'])
```

```python
df = df.reset_index(drop=True)
```

```python
columnas_numericas = ['poblacion_total', 'esperanza_vida', 'logro_educativo', 
                      'años_educacion', 'ingreso_percapita', 'idh', 
                      'idh_vida', 'idh_educacion', 'idh_ingreso']

for columna in columnas_numericas:
    df[columna] = pd.to_numeric(df[columna], errors='coerce')
```

```python
df = df.dropna(subset=columnas_numericas)
df = df.reset_index(drop=True)
```

```python
import matplotlib.pyplot as plt

df_departamentos = df[df['provincia_distrito'].isna()]

df_departamentos = df_departamentos.dropna(subset=['departamento', 'ingreso_percapita'])

df_departamentos['departamento'] = df_departamentos['departamento'].astype(str)

df_departamentos = df_departamentos.sort_values(by='ingreso_percapita', ascending=False)

# Gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(df_departamentos['departamento'], df_departamentos['ingreso_percapita'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Departamento')
plt.ylabel('Ingreso per cápita')
plt.title('Ingreso per cápita por Departamento')
plt.tight_layout()
plt.show()
```

```python
# Función para transformar ingreso per cápita a dólares
def convertir_a_dolares(ingreso_soles, tipo_cambio=3.8):
    return ingreso_soles / tipo_cambio
```

```python
df['ingreso_percapita_dolares'] = df['ingreso_percapita'].apply(convertir_a_dolares)

print(df[['departamento', 'ingreso_percapita', 'ingreso_percapita_dolares']].head())
```

```python
# Histograma

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df['ingreso_percapita_dolares'].dropna(), bins=15, color='lightgreen', edgecolor='black')
plt.title('Distribución del ingreso per cápita en dólares')
plt.xlabel('Ingreso per cápita (USD)')
plt.ylabel('Número de distritos/departamentos')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Gráfico de calor
variables = ['esperanza_vida', 'años_educacion', 'ingreso_percapita', 'idh', 'idh_vida', 'idh_educacion', 'idh_ingreso']

plt.figure(figsize=(10, 8))
sns.heatmap(df[variables].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de calor: correlación entre variables')
plt.tight_layout()
plt.show()
```

```python
# Gráfico de pie
df_poblacion = df.groupby('departamento')['poblacion_total'].sum().reset_index()

top5 = df_poblacion.sort_values('poblacion_total', ascending=False).head(5)

plt.figure(figsize=(8, 8))
plt.pie(top5['poblacion_total'], labels=top5['departamento'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 departamentos más poblados')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
