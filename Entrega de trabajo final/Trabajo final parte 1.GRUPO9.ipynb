1. Utilizando scipy. stats, crea un programa que modele la probabilidad de obtener un
número específico de éxitos en una serie de lanzamientos de una moneda sin trucos
(probabilidad de éxito p = 0. 5). Genera la probabilidad de obtener exactamente 7 caras
en 10 lanzamientos

```python
import math
import statistics
import numpy as np
from scipy import stats
import scipy
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom
!pip install openpyxl
n = 10 #Numero de lanzamientos
p = 0.5 #Probabilidad de éxito

#Probabiidad de obtener exactamente 7 caras 
#en una serie de lanzamientos de una moneda sin trucos

prob_cara_7 = binom.pmf(7, n, p)
print(f"La probabilidad de obtener exactamente 7 caras es:", round(prob_cara_7,4))
```

Calcula la probabilidad acumulada de obtener 7 o menos caras en 10 lanzamientos

```python
prob_acumulada_cara_7 = binom.cdf(7, n, p)
print(f"La probabilidad acumulada de obtener 7 o menos caras es :", round(prob_acumulada_cara_7,4))
```

Grafica la distribución binomial completa para los 10 lanzamientos, mostrando la
probabilidad de obtener desde 0 hasta 10 caras.

```python
#Los valores de éxitos posibles (de 0 a 10)
x = np.arange(0, n+1)

#Las probabiidades de cada número de éxitos
probs_x = binom.pmf(x, n, p)
#El gráfico
plt.bar(x, probs_x, color = "green", edgecolor = "black")
plt.xlabel("Número de caras")
plt.ylabel("Probabilidad")
plt.title("Distribución Binomial para los 10 lanzamientos, con probabilidad de éxito p=0.5")
plt.show()
```

Ahora la moneda tiene mayor probabilidad de caer en cara (probabilidad de éxito p =
0. 8). ¿Cómo cambian los resultados? Vuelve a realizar los ejercicios I, II y III.

```python
#Ahora la probabilidad cambió
p = 0.8
#Primer ejercicio con p=0.8
nuevo_prob_cara_7 = binom.pmf(7, n,p)
print(f"La nueva probabilidad de obtener exactamente 7 caras es:", round(nuevo_prob_cara_7,4))
```

```python
#Segundo ejercicio con p=0.8
nuevo_prob_acumulada_cara_7 = binom.cdf(7,n, p)
print(f"La nueva probabilidad acumulada de obtener 7 o menos cara es:", round(nuevo_prob_acumulada_cara_7,4))
```

```python
#Tercer ejercicio
#Los valores de éxitos posibles (de 0 a 10)
x = np.arange(0, n+1)

#Las nuevas probabiidades de cada número de éxitos
probs_x = binom.pmf(x, n, p)
#El gráfico
plt.bar(x, probs_x, color = "skyblue", edgecolor = "black")
plt.xlabel("Número de caras")
plt.ylabel("Probabilidad")
plt.title("La nueva distribución Binomial para los 10 lanzamientos, con probabilidad de éxito p=0.8")
plt.show()
```

Utilizando scipy. stats, realiza un análisis de una distribución normal con una media de
0 y desviación estándar de 1. Genera un conjunto de datos aleatorios con 1000
muestras a partir de esta distribución.

```python
#Los parámetros de la distrbución normal
media = 0
desviacion_estandar = 1
numero_muestras = 1000

#Toca generar el conjunto de datos con 1000 muestras a partir de la distirbución
datos = np.random.normal(media, desviacion_estandar, numero_muestras)
```

Calcula y muestra la media, la desviación estándar y la mediana de los datos
generados.

```python
media_datos = np.mean(datos)
desviacion_datos = np.std(datos, ddof=1)
mediana_datos = np.median(datos)

print(f"La media de los datos generados es:", round(media_datos,4))
print(f"La desviación estándar de los datos generados es:", round(desviacion_datos,4))
print(f"La mediana de los datos generados es:", round(mediana_datos,4))
```

Grafica el histograma de los datos y superpón la curva de la distribución normal
teórica.

```python
from scipy.stats import norm
#El histograma de los datos
plt.hist(datos, bins=11, density=True, alpha=0.6, color = "g", edgecolor= "black")

#Curva de la distribución normal teórica
x = np.linspace(min(datos), max(datos), 100)
y = norm.pdf(x, media, desviacion_estandar)
plt.plot(x, y, "r", linewidth = 2)

plt.xlabel("Valores")
plt.ylabel("Densidad de probabilidad")
plt.title("Histograma y curva de la distribución normal teórica")

plt.show()
```

Calcula la probabilidad de que un valor caiga dentro del rango de -1 a 1 (una desviación
estándar de la media).

```python
prob_ejercicio = norm.cdf(1, media, desviacion_estandar) - norm.cdf(-1, media, desviacion_estandar)
print(f"La probabilidad de que el valor caiga dentro del rango de -1 a 1 es:", round(prob_ejercicio,4))
```

```python
df_IDH =pd.read_excel("IDH_PNUD_2017.xlsx", sheet_name="Departamental", header=2)
def limpiar_datos(archivo):
    #Primero eliminamos las columnas cuyo título está vacío
    archivo = archivo.drop(columns=[col for col in archivo.columns if "Unnamed" in col]) 
    #Luego se borran las primeras 4 y últimas 3 filas sin datos
    archivo = archivo.iloc[4:].reset_index(drop=True)
    archivo = archivo.iloc[:-3].reset_index(drop=True)
    return archivo #Retorna el archivo limpio
df_IDH_final = limpiar_datos(df_IDH)
df_IDH_final
```

```python
#Gráfico 1
df_IDH_final = df_IDH_final.iloc[1:]
df_IDH_sorted_1 = df_IDH_final.sort_values("Esperanza de vida al nacer", ascending=True)

plt.figure(figsize=(10,8))
plt.barh(df_IDH_sorted_1["DEPARTAMENTO"], df_IDH_sorted_1["Esperanza de vida al nacer"], color="skyblue")

plt.xlabel("Esperanza de vida (años)")
plt.ylabel("Departamento")
plt.title("Esperanza de vida por departamento")

plt.show()
```
