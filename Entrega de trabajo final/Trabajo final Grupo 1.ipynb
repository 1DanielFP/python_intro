1. Utilizando scipy. stats, crea un programa que modele la probabilidad de obtener un
número específico de éxitos en una serie de lanzamientos de una moneda sin trucos
(probabilidad de éxito p = 0. 5). Genera la probabilidad de obtener exactamente 7 caras
en 10 lanzamientos

```python
from scipy.stats import binom

# Calcular la probabilidad de obtener exactamente 7 caras
probabilidad = binom.pmf(7, 10, 0.5)

print(f"La probabilidad de obtener exactamente 7 caras en 10 lanzamientos es: {probabilidad:.2f}")
```

Calcula la probabilidad acumulada de obtener 7 o menos caras en 10 lanzamientos

```python
#Calcular la probabilidad acumulada
probabilidad_acumulada = binom.cdf(7, 10, 0.5)
print(f"La probabilidad acumulada de obtener 7 o menos caras en 10 lanzamientos es: {probabilidad_acumulada:.2f}")
```

Grafica la distribución binomial completa para los 10 lanzamientos, mostrando la
probabilidad de obtener desde 0 hasta 10 caras.

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores posibles de éxito
x = np.arange(0, 11)

# Probabilidades de cada valor
probabilidades = binom.pmf(x, 10, 0.5)

# Gráfico
plt.bar(x, probabilidades)
plt.show()
```

Utilizando scipy. stats, realiza un análisis de una distribución normal con una media de
0 y desviación estándar de 1. Genera un conjunto de datos aleatorios con 1000
muestras a partir de esta distribución.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generar datos y realizar el gráfico
dat_al = norm.rvs(size=1000)
plt.hist(dat_al, bins=30, density=True, alpha=0.6, label='Datos')
plt.plot(np.linspace(-4, 4, 1000), norm.pdf(np.linspace(-4, 4, 1000)), 'r-', label='Teórica')
plt.legend()
plt.show()
```

Calcula y muestra la media, la desviación estándar y la mediana de los datos
generados.

```python
import numpy as np
from scipy import stats

# Simular datos aleatorios
np.random.seed(0)  # Para reproducibilidad
dat_al = np.random.normal(0, 1, 1000)

# Calcular la media, desviación estándar y la mediana
media = np.mean(dat_al).round(2)
desv_est = np.std(dat_al).round(2)
mediana = np.median(dat_al).round(2)

# Resultados
print(f"Media: {media}")
print(f"Desviación estándar: {desv_est}")
print(f"Mediana: {mediana}")
```

Grafica el histograma de los datos y superpón la curva de la distribución normal
teórica.

```python
plt.hist(dat_al, density=True, alpha=0.6, color='b', label='Datos simulados')

x = np.linspace(-3, 3, 1000)
y = norm.pdf(x, media, desv_est)
plt.plot(x, y, 'r-')

plt.xlabel('Valor')
plt.ylabel('Probabilidad')
plt.title('Histograma de datos aleatorios y distribución normal (media=0, sigma=1)')
plt.show()
```

Calcula la probabilidad de que un valor caiga dentro del rango de -1 a 1 (una desviación
estándar de la media).

```python
from scipy.stats import norm

# Calcular la probabilidad de que un valor caiga dentro del rango de -1 a 1 en una distribución normal estándar
probabilidad_rango = norm.cdf(1) - norm.cdf(-1)

# Resultados
print(f"Probabilidad de que un valor caiga en el rango de -1 a 1: {probabilidad_rango}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_excel('IDH_PNUD_2017-1.xlsx')  # Cambia la ruta si es necesario

# Limpieza
df.dropna(subset=["Número de Muertes maternas 1/", "Adolescentes madres o embarazadas"], inplace=True)
df["Adolescentes madres o embarazadas"] = df["Adolescentes madres o embarazadas"].str.replace(',', '.').astype(float)
df["Muertes_maternas_por_1000"] = df["Número de Muertes maternas 1/"] * 1000 / df["Adolescentes madres o embarazadas"]

# Estilo
sns.set(style="whitegrid")

# Gráfico 1: Muertes maternas por departamento
plt.figure(figsize=(12, 8))
sns.barplot(x="Número de Muertes maternas 1/", y="Departamento", data=df, palette="viridis")
plt.title("Muertes Maternas por Departamento")
plt.show()

# Gráfico 2: Adolescentes madres o embarazadas por departamento
plt.figure(figsize=(12, 8))
sns.barplot(x="Adolescentes madres o embarazadas", y="Departamento", data=df, palette="magma")
plt.title("Adolescentes Madres o Embarazadas por Departamento")
plt.show()

# Gráfico 3: Relación entre muertes maternas y adolescentes Embarazadas
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Adolescentes madres o embarazadas", y="Número de Muertes maternas 1/", data=df, hue="Departamento", palette="deep", s=100)
plt.title("Relación entre Muertes Maternas y Adolescentes Embarazadas")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Gráfico 4: Muertes maternas por 1000 adolescentes embarazadas
plt.figure(figsize=(12, 8))
sns.barplot(x="Muertes_maternas_por_1000", y="Departamento", data=df, palette="plasma")
plt.title("Muertes Maternas por 1000 Adolescentes Embarazadas")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos del excel
df = pd.read_excel('IDH_PNUD_2017-1.xlsx')

# Limpieza
df.dropna(subset=["Empoderamiento_Mujeres", "Mercado_Laboral_Mujeres"], inplace=True)
df["Brecha_Empoderamiento"] = df["Empoderamiento_Hombres"] - df["Empoderamiento_Mujeres"]
df["Brecha_Mercado_Laboral"] = df["Mercado_Laboral_Hombres"] - df["Mercado_Laboral_Mujeres"]

# Estilo
sns.set(style="whitegrid")

# Gráfico 1: Brecha de género en empoderamiento
plt.figure(figsize=(10, 6))
sns.barplot(x=df.index, y="Brecha_Empoderamiento", data=df, palette="coolwarm")
plt.title("Brecha de Género en Empoderamiento")
plt.xlabel("Índice")
plt.ylabel("Brecha (Hombres - Mujeres)")
plt.show()

# Gráfico 2: Brecha de género en el mercado laboral
plt.figure(figsize=(10, 6))
sns.barplot(x=df.index, y="Brecha_Mercado_Laboral", data=df, palette="viridis")
plt.title("Brecha de Género en el Mercado Laboral")
plt.xlabel("Índice")
plt.ylabel("Brecha (Hombres - Mujeres)")
plt.show()

# Gráfico 3: Relación entre empoderamiento y mercado laboral (Mujeres)
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Empoderamiento_Mujeres", y="Mercado_Laboral_Mujeres", data=df, hue="Tasa_Participacion_Mujeres", palette="magma", s=100)
plt.title("Relación entre Empoderamiento y Mercado Laboral (Mujeres)")
plt.xlabel("Empoderamiento (%)")
plt.ylabel("Mercado Laboral (%)")
plt.legend(title="Tasa Participación")
plt.show()

# Gráfico 4: Relación entre empoderamiento y mercado laboral (Hombres)
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Empoderamiento_Hombres", y="Mercado_Laboral_Hombres", data=df, hue="Tasa_Participacion_Hombres", palette="plasma", s=100)
plt.title("Relación entre Empoderamiento y Mercado Laboral (Hombres)")
plt.xlabel("Empoderamiento (%)")
plt.ylabel("Mercado Laboral (%)")
plt.legend(title="Tasa Participación")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_excel("IDH_PNUD_2017-1.xlsx")  # Asegúrate de que el archivo sea el correcto

# Filtrar datos de Crecimiento del PIB para Perú, Chile y Colombia
df_pib = df[df["País"].isin(["Perú", "Chile", "Colombia"])]

# Conversión de datos y limpieza
df_pib.dropna(subset=["Crecimiento PIB (%)"], inplace=True)
df_pib["Crecimiento PIB (%)"] = df_pib["Crecimiento PIB (%)"].astype(float)

# Estilo
sns.set(style="whitegrid")

# Gráfico 1: Comparación del Crecimiento del PIB (%) entre Perú, Chile y Colombia
plt.figure(figsize=(12, 6))
sns.lineplot(x="Año", y="Crecimiento PIB (%)", hue="País", data=df_pib, marker="o", palette="magma")

plt.title("Comparación del Crecimiento del PIB (%) entre Perú, Chile y Colombia")
plt.xlabel("Año")
plt.ylabel("Crecimiento PIB (%)")
plt.xticks(rotation=45)
plt.legend(title="País")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
