Grafica la distribución binomial completa para los 10 lanzamientos, mostrando la
probabilidad de obtener desde 0 hasta 10 caras.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

x = np.arange(0, 11)

y = binom.pmf(x, 10, 0.5)

plt.bar(x,y, color= "green")

plt.xlabel("Número de caras")
plt.ylabel("Probabilidad")
plt.title("Distribución Binomial (n=10, p=0.5)")
plt.xticks(x)
plt.show()
```

Utilizando scipy. stats, realiza un análisis de una distribución normal con una media de
0 y desviación estándar de 1. Genera un conjunto de datos aleatorios con 1000
muestras a partir de esta distribución.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm 

datos = norm.rvs(0, 1, size=1000)

plt.hist(datos, bins=np.linspace(-4,4,30), edgecolor='black', density=True)
plt.xlim(-4,4)
plt.show()
```

Calcula y muestra la media, la desviación estándar y la mediana de los datos
generados.

```python
import numpy as np
import scipy.stats as stats

muestra = stats.norm.rvs(loc=0, scale=1, size=1000) 

resumen = stats.describe(muestra)

promedio = np.round(np.mean(muestra), 2)
desv_std = np.round(np.std(muestra, ddof=1), 2)
valor_medio = np.round(np.median(muestra), 2)

print("=== Resultados Estadísticos ===")
print(f"→ Promedio: {promedio}")
print(f"→ Desviación estándar: {desv_std}")
print(f"→ Valor central (Mediana): {valor_medio}")
```

Grafica el histograma de los datos y superpón la curva de la distribución normal
teórica.

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 1000) 
y = stats.norm.pdf(x, loc=0, scale=1)  

plt.figure(figsize=(8, 5))
plt.hist(datos, bins=30, density=True, alpha=0.6, color='b', label='Histograma de datos')
plt.plot(x, y, 'r', label='Densidad teórica')

plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.title("Histograma de datos simulados vs. Distribución Normal Teórica")
plt.legend()
plt.show()
```
