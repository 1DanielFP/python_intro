```python
from scipy.stats import norm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import statistics
```

```python
df_distrib_normal =pd.DataFrame({'amount':data_distrib_normal})
```

```python

mean1 = df_distrib_normal['amount'].mean()
std1 = df_distrib_normal['amount'].std()
median1 =df_distrib_normal['amount'].median()

print(mean1)
print(std1)
print(median1)
```

# Emisión de CO2 en Brasil


```python
co2= pd.read_csv("C:/Users/heber/Documents/Python/51f7ab6a-24ed-4a06-9f5d-04873f1673fd_Data-1 (1).csv")
brasil_co2= co2[co2['Country Name'] == 'Brazil']
brasil_co2

#Ahora, seleccioniamos los indicadores de nuestro interes: Brasil y Co2 y nos quedamos solo con los años, ya que nos estamos enfocado en al evolución del CO2 en el tiempo. 
brasil_co2 = co2.loc[(co2['Country Name'] == 'Brazil') & (co2['Series Name'] == 'CO2 emissions (metric tons per capita)')]
brasil_co2 = brasil_co2.drop(columns=['Country Name', 'Country Code', 'Series Name', 'Series Code'])

#Ahora, luego de la selección, detallamos mejor los años. 
years = list(brasil_co2.columns)
for i in range(len(years)): years[i] = years[i][:4]
brasil_co2.columns = years

#Para evitar errores: 
brasil_co2 = brasil_co2.apply(pd.to_numeric, errors='coerce')


#Transponemos: 
tbrasil_co2 = brasil_co2.transpose()
tbrasil_co2.columns = ['CO2']





```

Ahora, procedemos a realizar los gráficos

# Evolución de emisión de C02 en Brasil

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(tbrasil_co2['CO2'], marker="o", color="black", label="Emisiones de CO2 en Brasil")

# Ajustar etiquetas del eje X sin necesidad de convertir los años a enteros
ax.set_xticks(range(0, len(tbrasil_co2), len(tbrasil_co2)//8))  # Seleccionar intervalos adecuados
ax.set_xticklabels(tbrasil_co2.index[::len(tbrasil_co2)//8], rotation=45)  # Usar índices originales como texto

ax.legend()
ax.set_title("Evolución de emisiones de CO2 en Brasil", fontdict={'fontsize': 18}, color='blue')
ax.set_xlabel("Año", fontdict={'fontsize': 16}, color='red')
ax.set_ylabel("Toneladas per cápita", fontdict={'fontsize': 16}, color='gray')
ax.grid()
plt.show()
```

# Número de camas ( Sudamerica y Norteamerica)

```python
df =  pd.read_csv("C:/Users/heber/Documents/Python/51f7ab6a-24ed-4a06-9f5d-04873f1673fd_Data-1 (1).csv")

#Agrupamos los países. 
sudamerica = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Paraguay', 'Peru', 'Uruguay', 'Venezuela']
norteamerica = ['Canada', 'United States', 'Mexico']
df_sudamerica = df[df['Country Name'].isin(sudamerica)]
df_norteamerica = df[df['Country Name'].isin(norteamerica)]
```

```python
df_sudamerica = df_sudamerica.rename(columns=lambda x: x.replace('YR', '') if 'YR' in x else x)
df_norteamerica = df_norteamerica.rename(columns=lambda x: x.replace('YR', '') if 'YR' in x else x)


df_sudamerica = df_sudamerica.melt(id_vars=['Country Name'], var_name='Year', value_name='Hospital Beds')
df_norteamerica = df_norteamerica.melt(id_vars=['Country Name'], var_name='Year', value_name='Hospital Beds')

#Evitamos errores
df_sudamerica['Hospital Beds'] = pd.to_numeric(df_sudamerica['Hospital Beds'], errors='coerce')
df_norteamerica['Hospital Beds'] = pd.to_numeric(df_norteamerica['Hospital Beds'], errors='coerce')

#Promedios
promedio_sudamerica = df_sudamerica.groupby('Year')['Hospital Beds'].mean()
promedio_norteamerica = df_norteamerica.groupby('Year')['Hospital Beds'].mean()
```

```python
#Gráfico: 

plt.figure(figsize=(10, 5))
plt.plot(promedio_sudamerica, label='Sudamérica', marker='o', linestyle='-')
plt.plot(promedio_norteamerica, label='Norteamérica', marker='s', linestyle='--')

plt.xlabel('Año')
plt.ylabel('Camas de Hospital por 1,000 habitantes')
plt.title('Evolución de Camas de Hospital en Sudamérica y Norteamérica')
plt.xticks(rotation=90)  
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Cargar los datos
idh_file = r"C:\Users\LENOVO\Downloads\IDH_PNUD_2017.xlsx"
global_data_file = r"C:\Users\LENOVO\Downloads\51f7ab6a-24ed-4a06-9f5d-04873f1673fd_Data-1 (1).csv"

# Cargar los DataFrames
df_idh = pd.read_excel(idh_file, sheet_name="Departamental", engine="openpyxl", header=2)
df_global = pd.read_csv(global_data_file, encoding="latin1")

# Función para limpiar datos
def procesar_datos(df):
    # Eliminar columnas sin nombre
    df = df.drop(columns=[col for col in df.columns if "Unnamed" in col])
    # Eliminar filas sin datos
    df = df.iloc[4:].reset_index(drop=True)
    df = df.iloc[:-3].reset_index(drop=True)
    return df

# Aplicar limpieza
df_idh = procesar_datos(df_idh)

# Limpiar nombres de columnas
df_idh.columns = df_idh.columns.str.strip()
df_global.columns = df_global.columns.str.strip()

# Mostrar las primeras filas para verificar la estructura de los datos
print(df_idh.head())
print(df_global.head())
print("Columnas en df_idh:", df_idh.columns.tolist())
```

```python
# Gráfico 1: Comparación de Esperanza de Vida por Departamento 
if "DEPARTAMENTO" in df_idh.columns and "Esperanza de vida al nacer" in df_idh.columns:
    df_idh_sorted_1 = df_idh.sort_values("Esperanza de vida al nacer", ascending=True)
    plt.figure(figsize=(12,6))
    plt.bar(df_idh_sorted_1["DEPARTAMENTO"], df_idh_sorted_1["Esperanza de vida al nacer"], color="red")
    plt.xticks(rotation=90, fontsize=10)
    plt.xlabel("Departamento", fontsize=12, color="black")
    plt.ylabel("Esperanza de vida (años)", fontsize=12, color="darkred")
    plt.title("Esperanza de vida por departamento", fontsize=14, fontweight='bold')
    plt.grid(axis="y", linestyle="--", linewidth=0.5)
    plt.show()
```
