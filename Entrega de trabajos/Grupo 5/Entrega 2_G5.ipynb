9. Haz un ```groupby``` que agrupe (`agg`) los datos por nivel de pobreza  y la nbi1. Queremos hallar el promedio y la mediana de los ingresos (inghog2d) y los gastos (gashog2d) por el producto cartesiano de las categorías de estas 2 variables - pobreza y nbi 1 -. 

El producto cartesiano es tan sólo todas las posibles combinaciones entre categorías. 
En este caso, tendremos:
- pobre extremo con nbi1
- pobre extremo sin nbi1
- pobre no extremo con nbi1
- pobre no extremo sin nbi1
- no pobre con nbi1
- no pobre sin nbi1

(Este es un  ejemplo de cómo, al cruzar un indicador de pobreza monetaria -pobreza- y pobreza multidimensional-nbis-, podemos seguir observando brechas).

````python
# Agrupar con el nivel de pobreza y nbi1, calculando el promedio y mediana de ingresos y gastos 
agg_pobreza_nbi = df_merged.groupby(['pobreza', 'nbi1']).agg(promedio_ingreso=('inghog2d', 'mean'), mediana_ingreso=('inghog2d', 'median'), promedio_gasto=('gashog2d', 'mean'), mediana_gasto=('gashog2d', 'median')).reset_index()
print(agg_pobreza_nbi)
````

3.2. Escribe una función que tome como input un diccionario, y que almacene en una lista a todos sus valores.  
Es decir. Si el diccionario es el de la clase: 
`ejem_dict =  {"Juana": "gallina", "Martin": "gato", "Carla": "conejo", "Cristina": "gato", "Diego": "perro" } `
Queremos un diccionario = `["gallina", "gato", "conejo","gato", "perro"]`

````python
def promedio(listadenumeros):
    '''
    Input:
    listadenumeros: lista de números (int o float)
    
    Output:
    Promedio: float
    '''
    return sum(listadenumeros) / len(listadenumeros)

# Definir la lista
lista = [1, 2, 3, 4, 5]

# Ejecutar
print(f'En base a la lista {lista}, el promedio es {promedio(lista)}')

#3.2 Diccionario
def diccionario3(diccionario):
    """
    Input:
    diccionario: dict

    Output:
    lista de valores: list
    """
    return list(diccionario.values())  

# Ejemplo
ejem_dict = {
    "Juana": "gallina",
    "Martin": "gato",
    "Carla": "conejo",
    "Cristina": "gato",
    "Diego": "perro"
}

resultado = diccionario3(ejem_dict)
print(f'El nuevo diccionario contiene {resultado}')
````

4. Tenemos la siguientes función: 

````python
def grado_escolar(edad):
    
    ''' calcula el grado escolar al que pertences segun tu edad
    Input: 
    edad : int
    output: tupla, compuestad de 
    grado: string
    esp: especifico que grado
    
    '''
    
    if ((edad < 6) | (edad > 17)):
        return "No está en edad escolar"
    #assert edad > 5 & edad < 17
    elif ((edad >=6) & (edad < 12)):
        grado = "Primaria"
        if edad == 6:
            esp = "1er grado"
        elif edad == 7:
            esp = "2do grado"
        elif edad == 8:
            esp = "3er grado"
        elif edad == 9:
            esp = "4to grado"
        elif edad == 10:
            esp = "5to grado"
        else:
            esp = "6to grado"
        
    else:
        grado = "Secundaria"
        if edad == 12:
            esp = "1er año"
        elif edad == 13:
            esp = "2do año"
        elif edad == 14:
            esp = "3er año"
        elif edad == 15:
            esp = "4to año"
        else:
            esp = "5to año"
            
    return grado, esp
````
